**Bab 1: Plot Paling Dasar (Line Plot)**

*1.1. Pengenalan Matplotlib dan Anatomi sebuah Plot*

Secara sederhana, **Matplotlib** adalah sebuah library (kumpulan kode) di Python yang berfungsi sebagai **peralatan menggambar utama** untuk membuat berbagai jenis grafik atau plot. Dengan Matplotlib, Anda bisa membuat berbagai macam visualisasi data yang statis, seperti:

- **Plot Garis (Line Plot)** untuk melihat tren dari waktu ke waktu.

- **Diagram Batang (Bar Chart)** untuk membandingkan jumlah antar kategori.

- **Histogram** untuk melihat distribusi atau sebaran data.

- **Scatter Plot** untuk melihat hubungan antara dua variabel.

Sebelum membuat plot, kita perlu tahu dua komponen utamanya:

- **Figure**: Ini adalah "kanvas" atau jendela keseluruhan tempat kita menggambar.

- **Axes**: Ini adalah area plot di dalam figure tempat data digambarkan, yang berisi sumbu-X dan sumbu-Y.

Untuk plot sederhana, Matplotlib akan membuatkannya secara otomatis untuk kita.

*1.2. Membuat Line Plot Sederhana*

```python
# Import modul pyplot dari matplotlib dengan alias plt
import matplotlib.pyplot as plt

# Data sederhana untuk sumbu X dan Y
tahun = [2018, 2019, 2020, 2021, 2022]
populasi = [100, 120, 150, 180, 210]

# Membuat line plot
plt.plot(tahun, populasi)

# Menampilkan plot yang sudah dibuat
plt.show()
```

*1.3. Memberi Judul dan Label*

```python
import matplotlib.pyplot as plt

tahun = [2018, 2019, 2020, 2021, 2022]
populasi = [100, 120, 150, 180, 210]

# Membuat line plot
plt.plot(tahun, populasi)

# Menambahkan judul pada plot
plt.title("Pertumbuhan Populasi")

# Menambahkan label pada sumbu X
plt.xlabel("Tahun")

# Menambahkan label pada sumbu Y
plt.ylabel("Jumlah Populasi (dalam juta)")

# Menampilkan plot
plt.show()
```

**Bab 2: Jenis-Jenis Plot Umum**

*2.1. Bar Chart (Diagram Batang)*

```python
import matplotlib.pyplot as plt

# Data kategorik
produk = ['A', 'B', 'C', 'D']
penjualan = [350, 500, 420, 600]

# Membuat bar chart
plt.bar(produk, penjualan)

# Menambahkan judul dan label
plt.title("Penjualan Produk")
plt.xlabel("Produk")
plt.ylabel("Jumlah Terjual")

# Menampilkan plot
plt.show()
```

*2.2. Histogram*

```python
import matplotlib.pyplot as plt

# Data numerik (contoh: nilai ujian)
data_nilai = [65, 72, 75, 80, 82, 85, 88, 90, 91, 95, 100]

# Membuat histogram dengan 5 kelompok (bins)
plt.hist(data_nilai, bins=5)

# Menambahkan judul dan label
plt.title("Distribusi Nilai Ujian")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi (Jumlah Mahasiswa)")

# Menampilkan plot
plt.show()
```

*2.3. Scatter Plot (Plot Sebar)*

```python
import matplotlib.pyplot as plt

# Data untuk dua variabel numerik
jam_belajar = [2, 3, 5, 7, 8]
nilai_ujian = [65, 70, 82, 88, 90]

# Membuat scatter plot
plt.scatter(jam_belajar, nilai_ujian)

# Menambahkan judul dan label
plt.title("Hubungan Jam Belajar dan Nilai Ujian")
plt.xlabel("Jam Belajar (dalam jam)")
plt.ylabel("Nilai Ujian")

# Menampilkan plot
plt.show()
```

**Bab 3: Kustomisasi Plot**

*3.1. Mengubah Warna, Gaya Garis, dan Penanda (Marker)*

```python
import matplotlib.pyplot as plt

tahun = [2018, 2019, 2020, 2021, 2022]
populasi = [100, 120, 150, 180, 210]

# Membuat plot dengan kustomisasi
plt.plot(tahun, populasi, color='red', linestyle='--', marker='o')

plt.title("Pertumbuhan Populasi")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Populasi (dalam juta)")
plt.show()
```

*3.2. Menambahkan Legenda*

```python
import matplotlib.pyplot as plt

# Data untuk dua kota
tahun = [2018, 2019, 2020, 2021, 2022]
populasi_a = [100, 120, 150, 180, 210]
populasi_b = [80, 85, 90, 100, 115]

# Plot data Kota A
plt.plot(tahun, populasi_a, marker='o', label='Kota A')

# Plot data Kota B
plt.plot(tahun, populasi_b, marker='x', label='Kota B')

plt.title("Pertumbuhan Populasi Kota A vs. Kota B")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Populasi (dalam juta)")

# Menampilkan legenda
plt.legend()

plt.show()
```

*3.3. Mengatur Ukuran Plot*

```python
import matplotlib.pyplot as plt

tahun = [2018, 2019, 2020, 2021, 2022]
populasi = [100, 120, 150, 180, 210]

# Mengatur ukuran figure (lebar, tinggi) dalam inci
plt.figure(figsize=(10, 5))

# Membuat plot seperti biasa
plt.plot(tahun, populasi, color='green', marker='s')

plt.title("Pertumbuhan Populasi")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Populasi (dalam juta)")
plt.show()
```

**Bab 4: Menampilkan Beberapa Plot (Subplots)**

```python
import matplotlib.pyplot as plt

# Data untuk plot pertama (Line Plot)
tahun = [2018, 2019, 2020, 2021, 2022]
populasi = [100, 120, 150, 180, 210]

# Data untuk plot kedua (Bar Chart)
produk = ['A', 'B', 'C', 'D']
penjualan = [350, 500, 420, 600]

# Membuat 1 baris, 2 kolom subplot
# Ini akan menghasilkan 2 area plot (axes) berdampingan
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# --- Menggambar di plot pertama (kiri, indeks 0) ---
ax[0].plot(tahun, populasi, color='blue', marker='o')
ax[0].set_title("Pertumbuhan Populasi")
ax[0].set_xlabel("Tahun")
ax[0].set_ylabel("Jumlah (dalam juta)")

# --- Menggambar di plot kedua (kanan, indeks 1) ---
ax[1].bar(produk, penjualan, color='green')
ax[1].set_title("Penjualan Produk")
ax[1].set_xlabel("Produk")
ax[1].set_ylabel("Jumlah Terjual")

# Menampilkan semua plot
plt.show()
```

*Soal #1: Dashboard Perbandingan Penjualan*

Buatlah sebuah visualisasi side-by-side (berdampingan) untuk menganalisis data penjualan. Plot pertama adalah line chart yang membandingkan tren penjualan produk A dan B selama 6 bulan. Plot kedua adalah bar chart yang menunjukkan total penjualan dari kedua produk tersebut.

```python
import matplotlib.pyplot as plt

# Data yang akan digunakan
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun']
penjualan_A = [200, 220, 250, 240, 260, 280]
penjualan_B = [180, 190, 210, 220, 230, 250]

# 1. Membuat figure dan axes untuk 2 plot (1 baris, 2 kolom)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# 2. Menggambar plot pertama (Line Chart)
ax[0].plot(bulan, penjualan_A, marker='o', label='Produk A')
ax[0].plot(bulan, penjualan_B, marker='x', label='Produk B')
ax[0].set_title('Tren Penjualan Bulanan')
ax[0].set_xlabel('Bulan')
ax[0].set_ylabel('Jumlah Terjual')
ax[0].legend()
ax[0].grid(True) # Menambahkan grid untuk kemudahan membaca

# 3. Menggambar plot kedua (Bar Chart)
total_penjualan = [sum(penjualan_A), sum(penjualan_B)]
nama_produk = ['Produk A', 'Produk B']
ax[1].bar(nama_produk, total_penjualan, color=['skyblue', 'salmon'])
ax[1].set_title('Total Penjualan per Produk')
ax[1].set_xlabel('Produk')
ax[1].set_ylabel('Total Terjual')

# 4. Menampilkan plot
plt.tight_layout() # Merapikan layout agar tidak tumpang tindih
plt.show()
```
